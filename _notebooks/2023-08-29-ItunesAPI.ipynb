## Javascript Itunes API
A search tool for Itunes using the Itunes API.

<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>iTunes Song Search</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            background-color: #f0f0f0;
            text-align: center;
        }
        #search-container {
            margin-top: 20px;
        }
        #search-input {
            padding: 10px;
            width: 300px;
            border: 1px solid #ccc;
            border-radius: 5px;
        }
        #search-button {
            padding: 10px 20px;
            background-color: #0073e6;
            color: white;
            border: none;
            border-radius: 5px;
            cursor: pointer;
        }
        #results-container {
            margin-top: 20px;
        }
    </style>
</head>
<body>
    <h1>iTunes Song Search</h1>
    <div id="search-container">
        <input type="text" id="search-input" placeholder="Enter a song or artist">
        <button id="search-button">Search</button>
    </div>
    <div id="results-container">
        <!-- Search results will appear here -->
    </div>
    <script>
        const searchInput = document.getElementById('search-input');
        const searchButton = document.getElementById('search-button');
        const resultsContainer = document.getElementById('results-container');
        searchButton.addEventListener('click', () => {
            const searchTerm = searchInput.value;
            if (searchTerm) {
                resultsContainer.innerHTML = '';
                fetch(`https://itunes.apple.com/search?term=${searchTerm}&entity=song`)
                    .then(response => response.json())
                    .then(data => {
                        data.results.forEach(song => {
                            const resultElement = document.createElement('div');
                            resultElement.innerHTML = `
                                <p><strong>${song.trackName}</strong> by ${song.artistName}</p>
                                <audio controls>
                                    <source src="${song.previewUrl}" type="audio/mpeg">
                                    Your browser does not support the audio element.
                                </audio>
                            `;
                            resultsContainer.appendChild(resultElement);
                        });
                    })
                    .catch(error => {
                        console.error('Error fetching data:', error);
                    });
            }
        });
    </script>
</body>
